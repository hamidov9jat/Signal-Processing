```python
import matplotlib.pyplot as pp
import numpy as np

ICE = np.genfromtxt(fname="ICE.txt",
                 delimiter="   ")

ICE

print(ICE)

# create 12 x 2 matrix for the year 1979
X = np.zeros([12, 2])
for j in range(12):
    X[j, 0] = ICE[0, 0] + (j+0.5)/12
    X[j, 1] = ICE[0, j+1]
print(X)

# create 504 = 12*42 lines and 2 columns
A = np.zeros([42*12, 2])
month = 0
for i in range(42):
    for j in range(12):
                    # time for month (j starts from 0)
                    # j+1 - 0.5
        A[month, 0] = ICE[i, 0] + (j+0.5)/12
        A[month, 1] = ICE[i, j+1]
        month += 1
    
        
print(A)

%matplotlib inline

pp.plot(A[:,0], A[:, 1])

import statistics as st

RAW = A

mu_RAW = st.mean(RAW[:,1])

print(mu_RAW)

stdev_RAW = st.pstdev(RAW[:,1])

print(stdev_RAW)

RAW

B = np.append(RAW, 
    np.array([[mu_RAW]]*504), axis=1)

np.array([[mu_RAW]]*3)

B

pp.plot(B[:,0],B[:,1],B[:,0],B[:,2])

B = np.append(B, 
    np.array([[mu_RAW + stdev_RAW]]*504), axis=1)

B

RAW

B = np.append(B, 
    np.array([[mu_RAW - stdev_RAW]]*504), axis=1)

B

pp.plot(B[:,0],B[:,1],B[:,0],B[:,2], B[:,0], B[:,3],
       B[:,0], B[:,4])

import pandas as pd

pd.Series.rolling(B[:,0],B[:,1],B[:,0],B[:,2], B[:,0], B[:,3],
       B[:,0], B[:,4])

df = pd.read_fwf('RAW_ICE.txt')

df.head()

```

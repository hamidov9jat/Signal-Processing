```python
url_az = "https://api.covid19api.com/live/country/azerbaijan/status/confirmed"
url_arm = "https://api.covid19api.com/live/country/armenia/status/confirmed"
url_tr = "https://api.covid19api.com/live/country/turkey/status/confirmed"

payload={}
headers = {}

response_az = rq.request("GET", url_az, headers=headers, data=payload).text
response_tr = rq.request("GET", url_tr, headers=headers, data=payload).text
response_arm = rq.request("GET", url_arm, headers=headers, data=payload).text

# For Az'}
response_info_for_country_az = json.loads(response_az) 
```

```python
# print(response)
print(response_info_for_country_az)
```

```python
country_list_statistics_az = []
for daily_info in response_info_for_country_az:
        # Format dates into y-m-d
        daily_info['Date'] = datetime.fromisoformat( daily_info['Date'][:-1] )
        daily_info['Date'] = daily_info['Date'].strftime('%Y-%m-%d')
        
        country_list_statistics_az.append([ daily_info["CountryCode"],
                            daily_info['Deaths'],
                            daily_info['Recovered'], daily_info['Active'], daily_info['Date']
        ])
```

```python
country_list_statistics_az
```

```python
country_df_az = pd.DataFrame(data=country_list_statistics_az, columns=['CountryCode','Deaths','Recovered','Active', 'Date'])
```

```python
country_df_az
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math
# if using a jupyter notebook
%matplotlib inline    
```

```python
f1=250
f2=500
f3 = 750

time = np.linspace(0, 2000, 500);

y1   = np.cos(2* np.pi * f1 * time)
y2 = np.cos(2 * np.pi * f2 * time)
y3 = np.cos(2 * np.pi * f3 * time)
# Give a title for the cosine wave plot

plt.title('Cosine wave')

# Give x axis label for the cosine wave plot

plt.xlabel('Time')

# Give y axis label for the cosine wave plot

plt.ylabel('Amplitude = cosine(time)')

 

# Draw the grid for the graph

plt.grid(True, which='both')

# y1   = np.cos(2* np.pi * 1/f1 * time)
# y2 = np.cos(2 * np.pi * 1/f2 * time)

plt.axhline(y=0, color='red')

plt.plot(time ,y1, time , y2,time, y3)
plt.show()
```
